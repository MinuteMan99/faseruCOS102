```python
BEGIN

    DECLARE Game1, Game2 AS STRING
    DECLARE Age1, Age2, Temp AS INTEGER

    OUTPUT "Enter the name of the first game:"
    INPUT Game1

    OUTPUT "Enter the age for the first game:"
    INPUT Age1

    OUTPUT "Enter the name of the second game:"
    INPUT Game2

    OUTPUT "Enter the age for the second game:"
    INPUT Age2

    Temp ← Age1
    Age1 ← Age2
    Age2 ← Temp

    OUTPUT "After swapping ages:"
    OUTPUT Game1, " Age: ", Age1
    OUTPUT Game2, " Age: ", Age2

END
```
