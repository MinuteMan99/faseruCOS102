```python
#simple interest
P = float(input("Enter principal amount: "))
R = float(input("Enter rate of interest (%): "))
T = float(input("Enter Time (years): "))

SI = (P * R * T) / 100

print("Simple interest =", SI)
```

```python
#compound interest
P = float(input("Enter principal amount: "))
R = float(input("Enter annual rate (%): "))
n = float(input("Enter number of items interest compounded per year: "))
t = float(input("Enter time (years): "))

R = R / 100

A = P * (1 + R/n) ** (n * t)
CI = A - P

print("Total amount =", A)
print("Compound interest=", CI)
```

```python
#annuity plan
PMT = float(input("Enter periodic payment amount: "))
R = R = float(input("Enter annual rate (%): "))
n = int(input("Enter time (years): "))
t = float(input("Enter time (years): "))

R = R / 100

A = PMT * (((1 + R/n) ** (n * t) / (R/n)))

print("Future Value of Annuity =", A)
```
